```python
def classify_email(email_text):
    """
    Classify customer email into categories: Billing, Technical Support, Feedback, or Others.
    
    Args:
        email_text (str): The customer email content
        
    Returns:
        str: Classification category
    """
    email_lower = email_text.lower()
    
    # Technical Support keywords
    technical_keywords = ['crash', 'error', 'bug', 'not working', 'issue', 'problem', 'help', 'fix']
    
    # Billing keywords
    billing_keywords = ['charge', 'invoice', 'payment', 'refund', 'bill', 'cost', 'price']
    
    # Feedback keywords
    feedback_keywords = ['suggestion', 'feedback', 'feature', 'improve', 'love', 'hate', 'opinion']
    
    if any(keyword in email_lower for keyword in technical_keywords):
        return "Technical Support"
    elif any(keyword in email_lower for keyword in billing_keywords):
        return "Billing"
    elif any(keyword in email_lower for keyword in feedback_keywords):
        return "Feedback"
    else:
        return "Others"

# Example usage
email = "My app crashes every time I try to log in. Can you help?"
category = classify_email(email)
print(f"Classification: {category}")
```
